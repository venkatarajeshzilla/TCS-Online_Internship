Checking missing values

```python
missing = df.isnull().sum()
missing[missing > 0]
```

```python
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.title("Missing Data Heatmap")
plt.show()
```

```python
df['Income'].fillna(df['Income'].median(), inplace=True)
```

```python
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

```python
sns.boxplot(x='Delinquent_Account', y='Credit_Utilization', data=df)
plt.title("Credit Utilization vs Delinquency")
plt.show()

```

```python
sns.countplot(x='Missed_Payments', hue='Delinquent_Account', data=df)
plt.title("Missed Payments Distribution")
plt.show()
```
